# [2015 Street Tree Census - Tree Data](https://dev.socrata.com/foundry/data.cityofnewyork.us/uvpi-gqnh)

```python
import pandas as pd
import numpy as np
from sodapy import Socrata
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
MyAppToken = 'insert_app_token_here'
```

```python
# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cityofnewyork.us", MyAppToken, 
                 username="insert_user_name_here", 
                 password="insert_password_here")
```

```python
# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("uvpi-gqnh", limit=683788)

# Convert to pandas DataFrame
tree_df = pd.DataFrame.from_records(results)
```

```python
tree_df.head()
```

```python
tree_df.shape
```

```python
tree_df.to_csv("../data/original_tree.csv",index=False)
```
